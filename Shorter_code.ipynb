```python
import time
import requests
from bs4 import BeautifulSoup

r = requests.get("https://tiki.vn/laptop/c8095?src=c.1846.hamburger_menu_fly_out_banner")
soup = BeautifulSoup(r.text, "html.parser")
```

```python
data_1 = []
data_2 = []
data_3 = []
data_4 = []
products = soup.find_all('div', class_ ='content')

for product in products:
    title = product.find('p', class_='title').text.strip()
    data_1.append(title)
    
    original = product.find('span', class_='price-regular').text.strip()
    data_2.append(original)
    
    sales = product.find('span', class_='sale-tag sale-tag-square').text.strip()
    data_3.append(sales)
    
    final = product.find('span', class_='final-price').text.strip()
    data_4.append(final)
```

```python
print(data_1)
```

```python
import pandas as pd

df = pd.DataFrame({'Product':data_1, 'Original':data_2, 'Sales':data_3, 'Final':data_4})
print(df)
```

```python
import time
import requests
from bs4 import BeautifulSoup 
import numpy as py
```

```python
pages = py.arange(1,16,1)
data_1 = []
data_2 = []
data_3 = []
data_4 = []

for page in pages:
    page = requests.get('https://tiki.vn/laptop/c8095?src=c.1846.hamburger_menu_fly_out_banner&page=' + str(page))
    soup = BeautifulSoup(page.text,'html.parser')
    products = soup.find_all('div', class_ ='content')
    time.sleep(2)
    
   
    
    for product in products:
        try:
            title = product.find('p', class_='title').text.strip()
            data_1.append(title)
        except:
            title = 'NA'
            data_1.append(title)
        
        try:
            original = product.find('span', class_='price-regular').text.strip()
            data_2.append(original)
        except:
            original = 'Na'
            data_2.append(original)
        
        try:
            sales = product.find('span', class_='sale-tag sale-tag-square').text.strip()
            data_3.append(sales)
        except:
            sales = ' '
            data_2.append(sales)
        
        try:
            final = product.find('span', class_='final-price').text.strip()
            data_4.append(final)
        except:
            final = ' '
            data_4.append(final)
```

```python
df_1 = pd.DataFrame({'Product':data_1, 'Original':data_2, 'Sales':data_3, 'Final':data_4})
print(df_1)
```
